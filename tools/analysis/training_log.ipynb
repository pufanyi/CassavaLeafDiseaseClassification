```python
log_file = "wandb_export_2024-11-19T14_09_10.066+08_00.csv"
```

```python
import pandas as pd

df = pd.read_csv(log_file)
```

```python
df.head()
```

```python
eval_loss = df[["train/global_step", "vit-large-patch16-384-wot - eval/loss", "vit-large-patch16-384 - eval/loss"]][df["train/global_step"] < 10000]
```

```python
# draw the loss curve
import matplotlib.pyplot as plt

plt.plot(eval_loss["train/global_step"], eval_loss["vit-large-patch16-384-wot - eval/loss"], label="Without Transformations")
plt.plot(eval_loss["train/global_step"], eval_loss["vit-large-patch16-384 - eval/loss"], label="With Transformations")
plt.legend()
plt.savefig("eval_loss_curve.pdf", bbox_inches='tight')
```
