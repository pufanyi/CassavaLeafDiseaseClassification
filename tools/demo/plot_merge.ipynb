```python
import numpy as np
import matplotlib.pyplot as plt

def plot_functions():
    # 创建数据点
    x = np.linspace(0, 1, 200)  # 增加点数使曲线更平滑
    
    # 计算函数值
    y1 = x
    y2 = (np.exp(x) - 1) / (np.exp(1) - 1)
    
    # 设置图形样式
    plt.style.use('seaborn')  # 使用更现代的样式
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # 绘制函数曲线
    ax.plot(x, y1, color='#2E86C1', linewidth=2, label='y = x')
    ax.plot(x, y2, color='#28B463', linewidth=2, label='y = (e^x-1)/(e-1)')
    
    # 绘制边界线
    ax.plot([1, 1], [0, 1], color='#E74C3C', linewidth=2, linestyle='--', label='x = 1')
    ax.plot([0, 1], [1, 1], color='#E74C3C', linewidth=2, linestyle='--')
    
    # 设置坐标轴
    ax.set_xlim(0, 1.1)
    ax.set_ylim(0, 1.1)
    
    # 添加网格和样式
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('y', fontsize=12)
    ax.set_title('函数比较图', fontsize=14, pad=15)
    
    # 优化图例位置和样式
    ax.legend(loc='lower right', framealpha=0.9, fontsize=10)
    
    # 添加坐标轴标签
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# 执行绘图
plot_functions()
```
